### Insert Values into each table

```python
query = """
--- Insert distinct values to sex table
insert into sex (sex)
select distinct sex from joined_table
where sex is not null;

--- Insert distinct values to age_groups table
insert into age_groups (age_group)
select distinct age_group from joined_table
where age_group is not null;

--- Insert distinct values to boroughs table
insert into boroughs (borough)
select distinct borough from joined_table
where borough is not null;


--- Insert distinct values to crime_categories table
insert into crime_categories (crime_category)
select distinct crime_category from joined_table
where crime_category is not null;


--- Insert and clean distinct jurisdictions table
insert into jurisdictions (jurisdiction_code)
select distinct jurisdiction_code
from joined_table
where jurisdiction_code is not null ;

--- Insert distinct values to law_codes table
insert into law_codes (law_code)
select distinct law_code from joined_table
where law_code is not null;

--- Insert distinct values to offense_levels table
insert into offense_levels (offense_level)
select distinct offense_level from joined_table
where offense_level is not null;


--- Insert distinct values to races table
insert into races (race)
select distinct race from joined_table
where race is not null;


--- Join foreign key and insert distinct values to crime_types table
insert into crime_types (crime_type, crime_category_id)
select distinct crime_type, cc.crime_category_id
from joined_table jt
left join crime_categories as cc on cc.crime_category = jt.crime_category
where crime_type is not null;

--- Join foreign keys and insert into arrests table
insert into arrests
	(arrest_id,
	arrest_date,
	crime_type_id,
	law_code_id,
	precinct_id,
	age_group_id,
	sex_id,
	offense_level_id,
	race_id,
	jurisdiction_id,
	longitude,
	latitude,
	borough_id)
select
	distinct jt.arrest_id,
	jt.arrest_date,
	ct.crime_type_id,
	lc.law_code_id,
	jt.precinct_id,
	ag.age_group_id,
	s.sex_id,
	ol.offense_level_id,
	r.race_id,
	j.jurisdiction_id,
	jt.longitude,
	jt.latitude,
	b.borough_id
from joined_table jt
left join sex s on jt.sex = s.sex
left join crime_categories cc on jt.crime_category = cc.crime_category
left join crime_types ct on cc.crime_category_id = ct.crime_category_id and ct.crime_type = jt.crime_type
left join law_codes lc on jt.law_code = lc.law_code
left join age_groups ag on jt.age_group = ag.age_group
left join offense_levels ol on jt.offense_level = ol.offense_level
left join races r on jt.race = r.race
left join jurisdictions j on jt.jurisdiction_code = j.jurisdiction_code
left join boroughs b on jt.borough = b.borough;

"""

cur.execute(query)

# commit the changes to the database
conn.commit()

print('Query successfully ran in PostgreSQL.')
```

```python
# Create a spark session and spark context
from pyspark.sql import SparkSession

#jarpath = "/Library/Frameworks/Python.framework/Versions/3.11/bin/postgresql-42.6.0.jar"
jarpath = "C:\\Users\\Jordan\\Joy\\crime-data\\postgresql-42.6.0.jar"
spark = SparkSession \
    .builder \
    .appName("NY Crime") \
    .config("spark.cores.max", "4") \
    .config('spark.executor.memory', '10G') \
    .config('spark.driver.maxResultSize', '10g') \
    .config('spark.kryoserializer.buffer.max', '512m') \
    .config("spark.driver.cores", "4") \
    .config("spark.jars", jarpath) \
    .getOrCreate()

# Configure the PostgreSQL JDBC driver path
# Must download PostgreSQL JDBC .jar file from https://jdbc.postgresql.org/download/
sc = spark.sparkContext

print("Using Apache Spark Version", spark.version)
```

```python
# Define the JDBC connection properties
hostname = "localhost"
port = "5432"
database = "ny_crime_5400"
username = "postgres"
password = "123"
jdbc_url = f"jdbc:postgresql://{hostname}:{port}/{database}"
jdbc_properties = {
  "user": username,
  "password": password,
  "driver": "org.postgresql.Driver"
}
print("Successfully configured to ny_crime_5400 database using PostgreSQL JDBC connection!")
```

#### Extract data from PostgresSQL to Spark


```python
# SQL query to join and select columns and filter out null values
query = """
SELECT a.arrest_date, cc.crime_category, a.precinct_id
FROM arrests AS a
left join crime_types as ct ON a.crime_type_id = ct.crime_type_id
left join crime_categories as cc ON ct.crime_category_id = cc.crime_category_id
WHERE cc.crime_category IS NOT NULL AND
      a.precinct_id IS NOT NULL AND
      a.arrest_date IS NOT NULL
"""

# Query from PostgreSQL Table
crime_sdf = spark.read \
    .format("jdbc") \
    .option("url", jdbc_url) \
    .option("driver", jdbc_properties["driver"]) \
    .option("query", query) \
    .option("user", jdbc_properties["user"]) \
    .option("password", jdbc_properties["password"]) \
    .load()

print("Query successfully ran and stored in crime_sdf")
```

```python
# View first five records
crime_sdf.limit(5).show()
```

```python
# View schema
crime_sdf.printSchema()
```

```python
print(f"Number of records in dataframe: {crime_sdf.count()}")
```

#### Store SDF and collect as array of Top 5 Precincts by Total Arrests to be referenced in form.html

```python
# Top 5 precincts with the highest arrests (Period: Jan 1, 2006 - Dec 31, 2022)
top5precincts_sdf = crime_sdf.groupBy("precinct_id").count().orderBy("count", ascending=False).limit(5)
top5precincts_sdf.show()
top5precincts = top5precincts_sdf.collect()
```

#### Store Max Date and Min Date to be referenced in form.html Title

```python
from pyspark.sql.functions import max, sum
```

```python
# Get Max date and Min date for Title to be referenced in form.html
max_arrest_date = crime_sdf.agg({"arrest_date": "max"}).collect()[0][0]
min_arrest_date = crime_sdf.agg({"arrest_date": "min"}).collect()[0][0]

print("The period of this dataset is:", min_arrest_date, " - ", max_arrest_date)
```

#### Store array of unique precincts to be referenced in the select list in form.html

```python
# Store array of unique precincts to be referenced in the select list in form.html
precinct_arr = crime_sdf.select("precinct_id").distinct().collect()
precinct_arr.sort()
print(f"Number of distinct precincts: {len(precinct_arr)}")
print(precinct_arr)
```

```python
# Import flask libraries
from flask import Flask, render_template, request
import findspark
findspark.init()

app = Flask(__name__)
```

```python
# Define Flask route to display a form for user inputs
@app.route("/", methods=["GET", "POST"])
def index():
    if request.method == "POST":
        # Get user inputs from form
        precinct_id = request.form["precinct_id"]

        # Filter crime_sdf based on user inputs
        filtered_sdf = crime_sdf.filter(crime_sdf.precinct_id == precinct_id)

        # Group by crime_category
        crimes_grouped_sdf = filtered_sdf.groupBy("crime_category").count().orderBy("count", ascending=False)

        # Convert Spark DataFrame to array
        crimes_grouped = crimes_grouped_sdf.collect()

        # Store the sum value as a row value
        sum_crimes_grouped = crimes_grouped_sdf.select(sum("count").alias("sum")).collect()[0]["sum"]

        # Render filtered data as an HTML table
        return render_template("index.html",
                               precinct_id = precinct_id,
                               crimes_grouped = crimes_grouped,
                               sum_crimes_grouped = sum_crimes_grouped,
                               mindate = min_arrest_date,
                               maxdate = max_arrest_date,)

    # Render form for user inputs

    return render_template("form.html",
                           precincts = precinct_arr,
                           mindate = min_arrest_date,
                           maxdate = max_arrest_date,
                           top5precincts = top5precincts)

```

```python
# Run flask app
if __name__ == "__main__":
    app.run(port=8011, debug=False)
```
